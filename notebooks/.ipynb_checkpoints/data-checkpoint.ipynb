1. Generate rates
2. Generate spikes

```python
import numpy as onp
from jax import numpy as np
from jax import random, vmap, jit
from jax.lax import scan, clamp
from jax.ops import index_min, index_update
from jax.numpy import clip

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette("colorblind")
sns.set_context("poster")
```

```python
dt = 1.0

def rate_step(params, yt, xt):
    """
    One step of OU process
    """
    dy = (params['mu'] - yt)/params['tau'] * dt + params['sigma'] * xt * np.sqrt(dt) 
    yt = yt + dy
    return yt, yt

def ou_process(params, time_steps, key):
    y0 = np.zeros((1, ))
    x = random.normal(key, shape=(time_steps, ))
    _, y = scan(lambda y, x: rate_step(params, y, x), y0, x)
    #y = index_min(y, y<0.0, 0.0)
    #y.at[y<0.0].set(0.0)
    y = np.clip(y, 0.0, y.max()) # Non-negative
    y = y * 0.1 / y.max() #  scale to [0, 1.0]
    y = onp.squeeze(y)
    spikes = random.bernoulli(key, p = y, shape = (neurons,) + y.shape) #(neurons, groups, time)
    return x, y, spikes
    
ou_process_jit = jit(vmap(ou_process, (None, None, 0)), (0,1))
```

```python
def split_keys(key, n_keys):
    new_keys = random.split(key, n_keys+1)
    return (k for k in new_keys[1:])
```

```python
params = {'mu': -2, 'tau': 20.0, 'sigma': 1.0} # dt in ms
neurons = 125
# Adust mean and tau to get 

seed = onp.random.randint(1e6)
key = random.PRNGKey(seed)
n_groups = 3
keys = random.split(key, n_groups+1)[1:]
time_steps = 300
xs, ys, spikes = ou_process_jit(params, time_steps, keys)
f, ax = plt.subplots(1,2, figsize=(10,4))
ax[0].plot(xs.T);
ax[1].plot(ys.T);
sns.despine()
```
