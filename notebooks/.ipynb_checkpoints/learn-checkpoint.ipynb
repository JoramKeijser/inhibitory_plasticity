# Simulate the feedforward network including ISP!

```python
import numpy as onp
from jax import numpy as np
from jax import random, vmap, jit
from jax.lax import scan, clamp
from jax.ops import index_min, index_update
from jax.numpy import clip

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette("colorblind")
sns.set_context("poster")
```

```python
# Parameters
p = {'vrest': -60.0, 'vth': -50.0, 'tref': 5.0,
         'gleak': 1.0, 've': 0.0, 'vi': -80.0,
         'te': 5.0, 'ti': 10.0, 'tau': 20.0, 
        't_stdp' : 20.0, 'rho': 5.0 * 10**(-3)}
dt = 1.0
```

Multiple neurons

```python
def make_exc_weights(Ne = 100):
    ge = 140 # pS
    K = np.arange(1, 9)
    P = 5
    Wgroups = 0.3 + 1.1 / (1 + (K - P)**4)
    W = np.repeat(Wgroups, Ne)
    eps = onp.random.uniform(0, 0.1, size=W.shape[0])
    W += eps
    return Wgroups, W
```

```python
def step(p, states, xt):
    """
    xt: exc. and inh spikes
    """
    v, spike, tspike, ge, gi, xe, xi, Wi, t = states # ((1, ); (1,), (Ne, ); (Ni, ))
    spike = np.where(v >= p['vth'], 1.0, 0.0)
    tspike = np.where(spike==1.0, t, tspike)
    print(xt[0].shape)
    eta = 1e-5
    alpha = 2 * p['rho'] * p['t_stdp'] 
    
    ge = (1 - dt / p['te']) * ge + xt[0]
    gi = (1 - dt / p['ti']) * gi + xt[1] # TO DO: Ni< Ne
    xe = (1 - dt / p['t_stdp']) * xe + spike # Postsynaptic trace
    xi = (1 - dt / p['t_stdp']) * xi + xt[1] # Presynaptic trace
    dvdt = (p['vrest'] - v) + 1/p['gleak'] * (np.dot(ge, p['We']) * 0.1 * (p['ve'] - v) + np.dot(gi, Wi) * (p['vi']- v) + 0)
    v = np.where(t - tspike <= p['tref'], p['vrest'], v + dt / p['tau'] * dvdt)
    # Update Weight - plasticity to the rescue
    Wi = np.where(spike == 1.0, Wi, Wi + eta * xi) # Update on post, using presynaptic trace
    Wi = np.where(xt[1] == 1.0, Wi, Wi + eta * (xe - alpha)) # Update on pre, using postsynaptic trace
    
    states =  v, spike, tspike, ge, gi, xe, xi, Wi, t+1
    return states, states

def gen_init_states(params, Wi = None):
    v = np.ones((1, )) * params['vrest']
    spike = np.zeros((1, )) * 1.0
    tspike = np.ones((1,)) * -np.inf
    ge = np.zeros((Ne * n_groups, )) * 1.0
    gi = np.zeros((Ne * n_groups, )) * 1.0
    xe = np.zeros((1, )) * 1.0
    xi = np.zeros((Ne * n_groups, )) * 1.0
    if Wi is None: # init to 0's
        Wi = np.zeros((Ne * n_groups, ))
    t = 0.0
    return v, spike, tspike, ge, gi, xe, xi, Wi, t

def lif_neuron(params, x, Wi = None):
    init_states = gen_init_states(params, Wi)
    _, states = scan(lambda s, x: step(params, s, x), init_states, x)
    return states

lif_neuron_jit = jit(lif_neuron, (0, ))
```

```python
Ne = 100
time_steps = 300
n_groups = 8
Wgroups, We = make_exc_weights(Ne)
p['We'] = We
```

# Multiple runs

```python
def make_data(key, time_steps = 1000):
    keys = random.split(key, n_groups+1)[1:]
    xs, ys, spikes = data_utils.ou_process_jit(params, time_steps, Ne * 2, keys) # Exc and inh.
    xe = spikes[:,:Ne].reshape((-1, time_steps)).T
    xi = spikes[:,Ne:].reshape((-1, time_steps)).T
    x = np.concatenate((xe[:,None], xi[:,None]), 1) #(time steps, 2, neurons)
    return x
```

```python
def split_keys(key, n_keys):
    new_keys = random.split(key, n_keys+1)
    return (k for k in new_keys[1:])
```

```python
n_trials = 100
seed = onp.random.randint(1e6)
key = random.PRNGKey(seed)
keys = split_keys(key, n_trials+1)

Wi_hist = []
alpha = 2 * p['t_stdp'] * p['rho']
for trial in range(n_trials):
    x = make_data(next(keys))
    if trial == 0: states = lif_neuron_jit(p, x)
    else: states = lif_neuron_jit(p, x, Wi[-1])
    v, spike, tspike, ge, gi, xe, xi, Wi,  t  = states
    if trial % 10 == 0: 
        print(trial, xe.mean(), alpha, spike.sum(), p['rho'] * 10**3)
        Wi_hist.append(Wi[-1])
```
